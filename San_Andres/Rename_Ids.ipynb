```python

Subjects=pd.read_excel('/home/ahaimovici/Documents/Teaching/Rectangle_Game/San_Andres/protocolo/lista_estudiantes.xlsx');      


where='meta10oct/';
files=[];
for file in listdir(where):
    if file.endswith(".csv"):
        files.append(file);

N_subjects=0;
useless=0;      
all_sequences=[];

Data=[];

for f in files:        
    N_correct=0;
    N_incorrect=0;
    Score=[0];        
    accum_score=0;
    nivel=0;
    
    if not('intdata' in f):            
        N_subjects+=1;
        df = pd.read_csv(where+f, delimiter=',');
        columns=df.columns.tolist();        
        tuples = [tuple(x) for x in df.values];                                       

        hit=0;
        false=0;
        miss=0;
        correct_reject=0;            
        correct=[];
        sequence=[];
        exp_time=0;
        Time_at_trial=[];
        time_flag=False;
        Level_seq=[];
        Hit_seq=[];
        False_seq=[];
        D_prime_seq=[];
        Criterion_seq=[];
        hit_norm=0;
        false_norm=0;
        trial_index=-1;
        max_nivel_time=99;


        Id=f[f.find('_')+1:f.find('2018')];
        
        
        grado=Subjects['Grupo'][Subjects['ID']==float(Id)].values[0];


        for row in tuples:                
            won=row[columns.index('won')];
            rt=row[columns.index('rt')];                
            trial_score=row[columns.index('trial_score')];                
            confident=row[columns.index('confident')];
            level=row[columns.index('level')];

            trial_type=row[columns.index('trial_type')];

            if isinstance(level,float) and not np.isnan(level):                                        
                Level_seq.append(float(level));
                trial_index+=1;

            if trial_type == 'circles': # start measuring time
                time_flag='True';

            if str.isdigit(str(rt)) and time_flag:
                exp_time+=int(rt)/60000;                                          

            if str(trial_score)!='nan':
                Score.append(Score[-1]+trial_score );   
                accum_score+=trial_score;
                if accum_score<0:
                    accum_score=0;
                elif accum_score>=10:
                    accum_score=accum_score-10;
                    nivel+=1;
                if nivel==10:
                    max_nivel_time=exp_time;
                
            if str(won)=='True':                                        
                N_correct+=1;
                correct=True;
                sequence.append(1);
            elif str(won)=='False':                    
                N_incorrect+=1;
                correct=False;     
                sequence.append(0);

            if str(confident)=='True' and correct:
                hit+=1; 
                hit_norm=hit/N_correct;
            elif str(confident)=='True' and not correct:
                false+=1;       
                false_norm=false/N_incorrect;
            elif str(confident)=='False' and correct:
                miss+=1;
            elif str(confident)=='False' and not correct:
                correct_reject+=1;

            if str(confident)=='True' or str(confident)=='False':                                    
                Hit_seq.append(hit_norm);
                Time_at_trial.append(exp_time);
                False_seq.append(false_norm);
                D_prime_seq.append(norm.ppf(hit_norm)-norm.ppf(false_norm));
                Criterion_seq.append(-0.5*(norm.ppf(hit_norm)+norm.ppf(false_norm)));

        stationary=-1;
        for j in range(0,min(len(Hit_seq),len(False_seq))):
            stationary+=1;
            if abs(Hit_seq[j]-Hit_seq[-1])/Hit_seq[-1] <0.05 and abs(False_seq[j]- False_seq[-1])/False_seq[-1] <0.05:
                break;


        '''
        if exp_time/60000 >5:
            print('----------------------------------------')          
            print('Id:',Id)
            print('grado:',grado)
            print('Time:',exp_time/60000)
        '''                
        N_trials=N_correct+N_incorrect;
        #print('Success:',sequence)
        #print('Level:',Level_seq)
        
        Data.append({'Id':Id,'Grado':grado,'Trials':N_trials,'Correct':N_correct/(N_correct+N_incorrect),
                    'High_confidence':(hit+false)/(N_trials),'hit':hit,'false':false,
                    'miss':miss,'correct_reject':correct_reject,
                     'Hit_seq':Hit_seq,'False_seq':False_seq,'Level_seq':Level_seq,'Stationary':stationary,
                     'Length_of_exp':exp_time,'Score':Score,'Time_at_trial':Time_at_trial,'nivel_maximo':nivel,
                     'max_nivel_time':max_nivel_time,'Success':sequence,'D_prime':D_prime_seq,'Criterion':Criterion_seq
                    })                        

Stationary=[];
Stationary_time=[];
counter=0;
for d in Data:
    if d['Level_seq'][1]!=0.5:
        print('Id:',d['Id'])
        print('grado:',d['Grado'])
        print('Trials:',d['Trials'])
        print('% correct:',d['Correct'])
        print('% high conf:',d['High_confidence'])
        print('Time:',d['Length_of_exp'])
        print('Hit:',d['hit'])
        print('False:',d['false'])
        print('Miss:',d['miss'])
        print('Correct Reject:',d['correct_reject'])
        print('Stationary:',d['Stationary'])
        print('Stationary Time:',d['Time_at_trial'][d['Stationary']])
        print('Nivel máximo:',d['nivel_maximo'])
        print('Tiempo Nivel máximo:',d['max_nivel_time'])
        
        if d['max_nivel_time']<5:
            counter+=1;
        
        
        Stationary.append(d['Stationary']);
        Stationary_time.append(d['Time_at_trial'][d['Stationary']]);
        
        plt.figure(1, figsize=(9, 3))
        plt.subplot(321)
        plt.plot(d['Level_seq'])
        plt.ylabel('level')
        plt.subplot(322)
        plt.plot(d['Time_at_trial'],d['Hit_seq'],d['Time_at_trial'],d['False_seq'],'r')  
        plt.legend(['hit','false'])
        plt.subplot(323)
        plt.plot(d['Score'],'-o')
        plt.subplot(324)
        plt.plot(np.diff(d['Score']),'o')
        plt.subplot(325)
        plt.plot(d['D_prime'])
        plt.subplot(326)
        plt.plot(d['Criterion'])
        plt.show()

plt.figure(2,figsize=(9,2))   
plt.subplot(1,2,1)
plt.hist(Stationary)
plt.xlabel('Stationary')
plt.ylabel('Counts')
plt.subplot(1,2,2)
plt.hist(Stationary_time)
plt.xlabel('Stationary time')
plt.ylabel('Counts')
plt.show()
                    
```
